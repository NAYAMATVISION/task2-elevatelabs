```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: for interactive graphs
import plotly.express as px

# For inline plots (if using Jupyter Notebook)
%matplotlib inline
```

```python
# Load the data
df = pd.read_csv(r'C:\Users\Nayamat\OneDrive\Desktop\elevatelabs\task2\Titanic-Dataset.csv')  

df.head()
```

```python
# Shape of the dataset
print("Shape of dataset:", df.shape)

# Data types and null values
df.info()

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Summary statistics for numerical columns
df.describe()

# Summary statistics for categorical columns
df.describe(include=['O'])
```

```python
# Mean, median, and standard deviation for selected numerical columns
print("Mean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Standard Deviation of Fare:", df['Fare'].std())
print("Skewness of Fare:", df['Fare'].skew())
print("Kurtosis of Fare:", df['Fare'].kurt())
```

```python
#age histogram
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='teal')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#boxplot of age
plt.figure(figsize=(8, 2))
sns.boxplot(x='Age', data=df, color='coral')
plt.title('Boxplot of Age')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Fare'], bins=40, kde=True, color='olive')
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 2))
sns.boxplot(x='Fare', data=df, color='skyblue')
plt.title('Boxplot of Fare')
plt.show()
```

```python
# Compute correlation for numeric columns only
corr_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived', palette='pastel')
plt.title('Survival Count by Gender')
plt.ylabel('Number of Passengers')
plt.show()

#checking survivak by gender 
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.violinplot(data=df, x='Survived', y='Age', palette='muted')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```
